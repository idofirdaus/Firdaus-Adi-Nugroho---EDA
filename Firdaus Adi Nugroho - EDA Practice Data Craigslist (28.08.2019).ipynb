# Handle Missing value

```python
#drop value that has more than 60% missing value

#vin have 1118215 missing values from 1723065 total data. we can conclude vin variabel have more than 60% missing value
#size have 1123967 missing values from 1723065 total data. we can conclude size variabel have more than 60% missing value
#we need to drop the vin and size collumn

data.drop(['vin', 'size'], axis =1, inplace=True)
```

```python
#make sure the deleted table
data.isnull().sum()
```

```python
#split the numeric and categoric
#we determine the numeric that consixt of integer or float type
#excluding the numeric, so we can catogorize named categorical

numeric = ['price', 'year', 'odometer', 'lat', 'long', 'weather']
categorical = data.drop(columns=numeric,axis=1)
numerical=data[numeric]
categorical.head()
```

```python
#doing imputation with the missing value

#all of categorical data impute to mode because mode won't change the type of data
#odometer imputes to median because odometer has skew data
#weather imputes to mean because the data relative stable.


#numeric
data['year'].fillna(numerical['year'].mode()[0], inplace=True)
data['odometer'].fillna(numerical['odometer'].median(), inplace=True)
data['weather'].fillna(data['weather'].mean(), inplace=True)

#categorical
data['manufacturer'].fillna(categorical['manufacturer'].mode()[0], inplace=True)
data['make'].fillna(categorical['make'].mode()[0], inplace=True)
data['condition'].fillna(categorical['condition'].mode()[0], inplace=True)
data['cylinders'].fillna(categorical['cylinders'].mode()[0], inplace=True)
data['fuel'].fillna(categorical['fuel'].mode()[0], inplace=True)
data['title_status'].fillna(categorical['title_status'].mode()[0], inplace=True)
data['transmission'].fillna(categorical['transmission'].mode()[0], inplace=True)
data['drive'].fillna(categorical['drive'].mode()[0], inplace=True)
data['type'].fillna(categorical['type'].mode()[0], inplace=True)
data['paint_color'].fillna(categorical['paint_color'].mode()[0], inplace=True)
data['image_url'].fillna(categorical['image_url'].mode()[0], inplace=True)
data['county_fips'].fillna(categorical['county_fips'].mode()[0], inplace=True)
data['county_name'].fillna(categorical['county_name'].mode()[0], inplace=True)
data['state_fips'].fillna(categorical['state_fips'].mode()[0], inplace=True)
data['state_code'].fillna(categorical['state_code'].mode()[0], inplace=True)
```

```python
#result action missiong value, if the all of number is 0, so there is no missing value.
data.isnull().sum()
```

# Perform Visualitation

## Correlation using Heatmap

```python
#create correlation with hitmap

#from this correlation, we know the value percentages of the correlation(1 or -1). The bigger value, so the correlation has big value too.
#weather has big correlation with the lattitude (-0,78). 

#create correlation
corr = data.corr(method = 'pearson')

#convert correlation to numpy array
mask = np.array(corr)

#to mask the repetitive value for each pair
mask[np.tril_indices_from(mask)] = False
fig, ax = plt.subplots(figsize = (15,12))
fig.set_size_inches(30,15)
sns.heatmap(corr, mask = mask, vmax = 0.9, square = True, annot = True)
```
